## Four methods to replace all the values:
```
1. apply()
2. map()
3. applymap()
4. replace()
```

### 1a. apply() on series object

````python
df['email'].apply(len)
````

````python
# Applying custom function 
def update_email(email):
    return email.upper()
````

````python
df['email'].apply(update_email)
````

````python
# To assign these values we will use assignment 
df['email'] = df['email'].apply(update_email)
df
````

````python
#Using lambda function to convvert it back to lower case
df['email'] = df['email'].apply(lambda x: x.lower())
df
````
